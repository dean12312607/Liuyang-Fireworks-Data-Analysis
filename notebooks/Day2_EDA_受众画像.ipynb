```python
# ==========================================
# 模块：02_EDA_and_Visualization (第一小节终极版：渠道与交叉特征)
# 作用：处理多选题、绘制交叉分析热力图，产出高阶业务洞察
# ==========================================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['font.sans-serif'] = ['SimHei']  
plt.rcParams['axes.unicode_minus'] = False 

# 1. 读取我们的带中文标签的 Excel 原文件并清洗废卷
df_text = pd.read_excel(r"C:\Users\wyt15\Desktop\市调第2次\原始数据（文本）.xlsx")
df_text['所用时间_数值'] = df_text['所用时间'].str.replace('秒', '').astype(int)
df_eda = df_text[df_text['所用时间_数值'] >= 60].copy()

# 2. 准备一张更大的双图画布
plt.figure(figsize=(16, 7))

# ---- 图3：多选题硬核处理：认知渠道排行榜 ----
plt.subplot(1, 2, 1) 
# 面试官看点：用代码精妙处理含有特殊符号（┋）的多选题
channels = df_eda['15、您主要是通过哪些途径接触和了解到浏阳烟花的？'].dropna().astype(str)
# 按“┋”拆分并炸开（explode）成单行，然后统计数量
channel_counts = channels.str.split('┋').explode().value_counts().head(8) 

# 画水平柱状图
sns.barplot(y=channel_counts.index, x=channel_counts.values, palette='Blues_r')
plt.title('图3：受众获取浏阳烟花信息的渠道TOP8', fontsize=16, fontweight='bold')
plt.xlabel('被提及的总次数 (频次)', fontsize=12)
plt.ylabel('信息获取途径', fontsize=12)

# ---- 图4：高阶交叉分析：不同年龄段对演出形式的偏好 ----
plt.subplot(1, 2, 2)
# 面试官看点：使用 Pandas 交叉表（crosstab）直接计算百分比
cross_tab = pd.crosstab(df_eda['1、您的年龄：'], df_eda['13、您更喜欢哪种浏阳烟花演出形式'], normalize='index') * 100

# 画热力图，颜色越深代表比例越高
sns.heatmap(cross_tab, annot=True, fmt='.1f', cmap='YlGnBu', cbar_kws={'label': '该年龄段内的占比 (%)'})
plt.title('图4：不同年龄段对烟花演出形式的偏好热力图', fontsize=16, fontweight='bold')
plt.ylabel('受众年龄段', fontsize=12)
plt.xlabel('偏好的演出形式', fontsize=12)
plt.xticks(rotation=15) # 防止底部文字挤在一起

# 调整布局，展示图表
plt.tight_layout()
plt.savefig(r"C:\Users\wyt15\Desktop\市调第2次\output\图片3+4.png", dpi=300, bbox_inches='tight')
print("✅ 图片已自动保存至桌面 output 文件夹！")
plt.show()
# 确保你已经建立了 output 文件夹，如果没有，请先在桌面文件夹里手动新建一个
```

```python
# ==========================================
# 模块：02_EDA_and_Visualization (第一小节收官：核心漏斗模型)
# 作用：计算从“知晓”到“付费”的转化率漏斗，定调受众消费现状
# ==========================================

import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei']  
plt.rcParams['axes.unicode_minus'] = False 

# 1. 读取数据并清洗废卷
df_text = pd.read_excel(r"C:\Users\wyt15\Desktop\市调第2次\原始数据（文本）.xlsx")
df_text['所用时间_数值'] = df_text['所用时间'].str.replace('秒', '').astype(int)
df_eda = df_text[df_text['所用时间_数值'] >= 60].copy()

# 2. 核心数据逻辑：计算漏斗每一层的人数
# 层级1：潜在受众（有效调研总人数）
funnel_1 = len(df_eda)

# 层级2：知晓受众（对烟花有一定了解，排除了完全不了解的人）
funnel_2 = df_eda[df_eda['11、您对于浏阳烟花是否了解：'].isin(['非常了解', '比较了解', '一般'])].shape[0]

# 层级3：体验受众（曾经线下实地观看过烟花的人）
funnel_3 = df_eda[df_eda['8、您是否线下观看过浏阳烟花'] == '是'].shape[0]

# 层级4：核心付费受众（明确表示同意或非常同意参与付费活动）
funnel_4 = df_eda[df_eda['20、您愿意参与付费的浏阳烟花相关活动'].isin(['同意', '非常同意'])].shape[0]

values = [funnel_1, funnel_2, funnel_3, funnel_4]
labels = [
    '1. 潜在受众 (调研总样本)', 
    '2. 知晓受众 (了解浏阳烟花)', 
    '3. 体验受众 (曾线下观看)', 
    '4. 核心受众 (愿意付费参与)'
]

# 3. 开始绘制高颜值的漏斗图 (利用左右对称的水平柱状图实现)
plt.figure(figsize=(10, 6))
# 颜色从浅蓝过渡到深蓝，代表人群的不断沉淀和精准化
colors = ['#87CEFA', '#6495ED', '#4169E1', '#0000CD']

# 计算让柱子居中的左侧缩进量（这是画漏斗图的代码小技巧）
max_val = max(values)
lefts = [(max_val - v) / 2 for v in values]

# 反转顺序画图，让最大的漏斗口在最上面
plt.barh(labels[::-1], values[::-1], left=lefts[::-1], color=colors[::-1], height=0.6)

# 4. 给漏斗中心打上转化率的烙印
for i, v in enumerate(values[::-1]):
    # 计算当前层对第一层总样本的转化率
    rate = int(v / values[0] * 100)
    plt.text(max_val/2, i, f"{v}人 (转化留存: {rate}%)", 
             ha='center', va='center', color='white', fontweight='bold', fontsize=12)

# 5. 图表美化与展示
plt.title('图5：受众核心消费现状转化漏斗', fontsize=18, fontweight='bold', pad=20)
plt.axis('off') # 关闭多余的坐标轴黑框，让图表充满高级感
plt.savefig(r"C:\Users\wyt15\Desktop\市调第2次\output\现状漏斗图.png", dpi=300, bbox_inches='tight')
print("✅ 图片已自动保存至桌面 output 文件夹！")
plt.show()
```
