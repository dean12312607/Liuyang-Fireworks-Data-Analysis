```python
# ==========================================
# 步骤：03_Handling_Missing_Values (处理缺失值-双轨制策略)
# ==========================================

# 1. 删掉全为空值、毫无用处的“来源详情”列
df_clean = df_clean.drop(columns=['来源详情'])

# =======================================================
# 【数据集 A】全量数据集：用于第一节“受众画像与漏斗分析”
# 保留那 160 个不了解的人，挖掘潜在受众
# =======================================================
df_all = df_clean.copy()
df_all.to_csv('浏阳烟花_全量数据_用于第一节画像.csv', index=False, encoding='utf-8-sig')
print(f"✅ 【数据集 A】已保存: '浏阳烟花_全量数据_用于第一节画像.csv' (共 {df_all.shape[0]} 份)")
print("   -> 作用：用于挖掘未触达人群的特征，画出最完整的用户画像。\n")

# =======================================================
# 【数据集 B】核心评价数据集：用于第二节“Logistic回归建模”
# 剔除因为“不了解”而没有进行满意度打分的样本
# =======================================================
# 只要“14、您对浏阳烟花的满意程度—造型”这一列是空的，我们就删掉这行
df_core = df_clean.dropna(subset=['14、您对浏阳烟花的满意程度—造型']).copy()
df_core.to_csv('浏阳烟花_核心数据_用于第二节建模.csv', index=False, encoding='utf-8-sig')
print(f"✅ 【数据集 B】已保存: '浏阳烟花_核心数据_用于第二节建模.csv' (共 {df_core.shape[0]} 份)")
print("   -> 作用：消除了空值，专供带入统计模型计算边际效应，绝不会报错。")
```
